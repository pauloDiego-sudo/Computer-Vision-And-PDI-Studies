```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('farol.jpg')
```

```python
type(img)
img.shape
```

## Como Exibir imagem

```python
cv2.imshow('Imagem',img)
cv2.waitKey(0)
```

### Pelo matplotlib

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imwrite('farolR.jpg',img_rgb[:,:,0])
cv2.imwrite('farolG.jpg',img_rgb[:,:,1])
cv2.imwrite('farolB.jpg',img_rgb[:,:,2])
plt.imshow(img_rgb)
plt.show()
```

```python
plt.imshow(img_rgb[:,:,0])
plt.show()
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,'gray')
plt.show()
cv2.imwrite('farolgray.jpg',img_gray)
```
